```python
import numpy as np
import layercake as lc
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
X = np.random.normal(0, 1, [5, 2])
y = np.sum(X, axis=-1, keepdims=True)
data_set = {'X': X, 'y': y}
ds, val = [lc.DataSource(d, 1) for d in lc.split_data(data_set, [.8, .2])]
# ds = lc.DataSource(data_set, 1)
ds, val, _ = [lc.DataSource(d, 20) for d in lc.load_mnist()]
```

```python
input_layer = lc.InputLayer(784)
model = lc.Sequential(input_layer)
model.add(lc.NeuronLayer(128))
model.add(lc.NonLinearity('relu'))
model.add(lc.NeuronLayer(10))

model.compile(
#     optimizer=lc.Optimizer(learning_rate=0.01),
    optimizer=lc.MomentumOptimizer(momentum=0.9, learning_rate=0.0031),
#     optimizer=lc.L2Optimizer(learning_rate=0.03),
    loss_layer=lc.CategoricalLossLayer()
)
```

```python
model.train(data_source=ds, with_validation=val, num_epochs=500, tolerance=1.02)
```

```python
print(len(model.val_losses))
```

```python
%matplotlib inline
plt.figure(figsize=[16,10])
plt.plot(np.asarray(model.val_losses))
plt.show()
```

```python
plt.figure(figsize=[16,16])
plt.plot(np.asarray(model.train_losses))
plt.show()
```

```python
plt.figure(figsize=[16,16])
plt.plot(np.asarray(model.tl))
plt.show()
```
